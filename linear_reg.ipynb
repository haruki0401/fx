```python
data_dir = "./"
data = pd.read_csv(data_dir + "USD_JPY_2009-2019.csv")
```

```python
data.head()
```

```python
data2 = np.array(data)
```

```python
day_ago = 25

X = np.zeros((len(data2), day_ago))

for i in range(0,day_ago):
    X[i:len(data2),i] = data2[0:len(data2)-i,4]
```

```python
data_show = pd.DataFrame(X)
data_show
```

```python
Y = np.zeros(len(data2))
pre_day = 1
Y[0:len(Y)-pre_day] = X[pre_day:len(X),0] - X[0:len(X)-pre_day,0]
```

```python
data_show = pd.DataFrame(Y)
data_show
```

```python
origin_X = np.copy(X)
tmp_mean = np.zeros(len(X))

for i in range(day_ago,len(X)):
    tmp_mean[i] = np.mean(origin_X[i-day_ago+1:i+1,0])
    for j in range(0,X.shape[1]):
        X[i,j] = (X[i,j] - tmp_mean[i])
    Y[i] = Y[i] #変化なし 
    
```

```python
data_show = pd.DataFrame(X)
data_show
```

```python
X_train = X[200:2332,:]
Y_train = Y[200:2332]
            
X_test = X[2332:len(X)-pre_day,:]
Y_test = Y[2332:len(X)-pre_day]
```

```python
from sklearn import linear_model
linear_reg_model = linear_model.LinearRegression()
```

```python
linear_reg_model.fit(X_train,Y_train)
```

```python
print(linear_reg_model.intercept_)
print(linear_reg_model.coef_)
```

```python
Y_pred = linear_reg_model.predict(X_test)
```

```python
result = pd.DataFrame(Y_pred)
result.columns = ['Y_pred']
result['Y_test'] = Y_test
```

```python
sns.set_style('darkgrid')
sns.regplot(x='Y_pred',y='Y_test',data=result)
```

```python
correct_num = 0
for i in range(len(Y_pred)):
    if Y_pred[i] * Y_test[i] >= 0:
        correct_num += 1
        
print("day : "+str(len(Y_pred))+" correct_num : "+str(correct_num)+" correct percentage : "+str(correct_num/len(Y_pred)*100))
```

```python
sum = 0

for i in range(len(Y_test)):
    if Y_pred[i] >= 0:
        sum += Y_test[i]
    else:
        sum -= Y_test[i]
        
print("sum : %1.3lf" %sum)
```

```python
total_return = np.zeros(len(Y_test))

if Y_pred[i] >=0: # the 1st day
    total_return[0] = Y_test[i]
else:
    total_return[0] = -Y_test[i]

for i in range(1, len(result)): # from the 2nd day
    if Y_pred[i] >=0:
        total_return[i] = total_return[i-1] + Y_test[i]
    else:
        total_return[i] = total_return[i-1] - Y_test[i]

plt.plot(total_return)
```
